```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import spacy
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the BBC data 
```

```python
bbc_data = pd.read_csv('bbc_news.csv')
```

```python
bbc_data.head()
```

```python
bbc_data.info()
```

```python
titles = pd.DataFrame(bbc_data['title'])
```

```python
titles.head() 
```

```python
# Clean the data
# - lower case
# - remove stopwords
# - remoe punctuation 
# - tokenize
# - lemmatize 
```

```python
titles['lowercase'] = titles['title'].str.lower() # covert the titles to lower case and add them to a new series in the DF
titles.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
titles['no_stopwords'] = titles['lowercase'].apply( # create a new series in DF call 'stopwords' from lowercase
    lambda x: ' '.join(
        [word for word in x.split() if word not in (en_stopwords)] # The values in the new series should be sentences without stopwords
    )
)

titles.head()
```

```python
titles['no_stopwords_no_punkt'] = titles.apply(lambda x: re.sub(r"[^\w\s]", "", x['no_stopwords']), axis=1)

titles.head()
```

```python
titles['tokens_raw'] = titles.apply(lambda x: word_tokenize(x['title']), axis=1)
titles['tokens_cleanr'] = titles.apply(lambda x: word_tokenize(x['no_stopwords_no_punkt']), axis=1)

titles.head()                                     
```

```python
lemmatizer = WordNetLemmatizer()
titles['token_clean_lemmatized'] = titles['tokens_cleanr'].apply(
    lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])

titles.head()
```

```python
tokens_raw_list = sum(titles['tokens_raw'], [])
tokens_clean_list = sum(titles['token_clean_lemmatized'], [])

print(tokens_raw_list[:10])

print(tokens_clean_list[:10])
```

```python
# POS tagging
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
spacy_doc = nlp(' '.join(tokens_raw_list)) 

# creates a single string with spaces between the words. 
# nlp process the string linguistically, and then stores it in a spacy doc object
```

```python
print(spacy_doc[:20])
```

```python
pos_df = pd.DataFrame(columns=['token', 'pos_tag']) # Create a new DF with two columns
```

```python
for token in spacy_doc: 
    pos_df = pd.concat([pos_df, # Add the new row that was created to the pos_df 
                   pd.DataFrame.from_records( # Creates a one row dataframe from the following values
                       [{'token' : token.text, # Get the word
                         'pos_tag' : token.pos_}])], # Get the POS tag
                  ignore_index=True)
                  
```

```python
# Create a new DF
# - Group the data by two colums
# - size() counts how often the 'token' + 'pos_tag' pair occurs
# - reset_index() resets the table so that 'token', 'pos_tag' and 'counts' all have their own column - with 'counts' being the new column name

pos_df_counts = pos_df.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts',ascending=False)
```

```python
pos_df_counts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == 'NOUN']
print(nouns[:10])
```

```python
verbs = pos_df_counts[pos_df_counts.pos_tag == 'VERB']
print(verbs[:10])
```

```python
adj = pos_df_counts[pos_df_counts.pos_tag == 'ADJ']
print(adj[:10])
```

```python
spacy_doc_clean = nlp(' '.join(tokens_clean_list)) 
pos_df_2 = pd.DataFrame(columns=['token', 'pos_tag'])

for token in spacy_doc_clean: 
    pos_df_2 = pd.concat([pos_df_2, # Add the new row that was created to the pos_df 
                   pd.DataFrame.from_records( # Creates a one row dataframe from the following values
                       [{'token' : token.text, # Get the word
                         'pos_tag' : token.pos_}])], # Get the POS tag
                  ignore_index=True)

pos_df_counts_2 = pos_df_2.groupby(['token', 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts',ascending=False)

pos_df_counts_2.head(10)
```

```python
nouns_2 = pos_df_counts_2[pos_df_counts_2.pos_tag == 'NOUN']
print(nouns_2[:10])
```

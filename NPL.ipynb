```python
customer_reviews = [
    "sam was a great help to me in the store",
    "the casier was very rude to me in store. I think her name was eleanor",
    "amazing work from sadeen",
    "sara was able to help me find the item's I needed quickly",
    "lucy is a great addition to the team",
    "great service from sara, she found me what I wanted"
]
```

```python
sarahs_reviews = []
pattern_to_find = r"sarah?" # ? means the preceeding character is optional
```

```python
sarahs_reviews = [
    string for string in customer_reviews 
    if re.search(pattern_to_find, string)]
print(sarahs_reviews)
```

```python
sarahs_reviews
```

```python
a_reviews = []
pattern_to_match = r"^a" # ^ means look for strings starting with the letter after
```

```python
a_reviews = [
    string for string in customer_reviews 
    if re.search(pattern_to_match, string)
]
print(a_reviews)
```

```python
y_reviews = [
    string for string in customer_reviews
    if re.search(r"y$", string) # $ finds strings ending in a letter
]
print(y_reviews)
```

```python
needwant_reviews = []
pattern_to_search = r"(need|want)ed" # | acts as a 'or' operator
```

```python
needwant_reviews = [
    string for string in customer_reviews
    if re.search(pattern_to_search, string)
]
print(needwant_reviews)
```

```python
no_punc_reviews = [
    re.sub(r"[^\w\s]", " ", string) 
    for string in customer_reviews # ^ means not, \w word characters, \s white space characterrs
]
print(no_punc_reviews)

# Note for loop conditionals [ Value/Experession to return, Loop, any conditions] 
```

```python
# N-gram = a sequence of N-tokens
# N=1 unigrams ('i', 'love' 'nlp')
# N=2 bigrams ('i love', 'love nlp')
# N=3 trigram ('i love nlp')
```

```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tokens = ['the', 'rise', 'of', 'artificial', 'intelligence', 'has', 'led', 'to', 'significant', 'advancements', 'in', 'natural', 'language', 'processing', 'computer', 'vision', 'and', 'other', 'fields', 'machine', 'learning', 'algorithms', 'are', 'becoming', 'more', 'sophisticated', 'enabling', 'computers', 'to', 'perform', 'complex', 'tasks', 'that', 'were', 'once', 'thought', 'to', 'be', 'the', 'exclusive', 'domain', 'of', 'humans', 'with', 'the', 'advent', 'of', 'deep', 'learning', 'neural', 'networks', 'have', 'become', 'even', 'more', 'powerful', 'capable', 'of', 'processing', 'vast', 'amounts', 'of', 'data', 'and', 'learning', 'from', 'it', 'in', 'ways', 'that', 'were', 'not', 'possible', 'before', 'as', 'a', 'result', 'ai', 'is', 'increasingly', 'being', 'used', 'in', 'a', 'wide', 'range', 'of', 'industries', 'from', 'healthcare', 'to', 'finance', 'to', 'transportation', 'and', 'its', 'impact', 'is', 'only', 'set', 'to', 'grow', 'in', 'the', 'years', 'to', 'come']
print(tokens)
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens, 1)).value_counts()) # pd.Series is like a single column of data in a spreadsheet
print(unigrams[:10])
```

```python
unigrams[:20].sort_values().plot.barh(
    color="lime", 
    width=0.9, 
    figsize=(12,8)
)
plt.title("20 most occuring unigrams")
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(bigrams[:10])
```

```python
trigrams = (pd.Series(nltk.ngrams(tokens, 3)).value_counts())
print(trigrams[:10])
```
